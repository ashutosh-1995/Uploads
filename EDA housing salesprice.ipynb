Lets create new column to classify this data even more

```python
h['Year']= 0
```

```python
h.loc[((h['YearBuilt']>= 1870) & (h['YearBuilt']< 1880)), 'Year']= '1870s'
h.loc[((h['YearBuilt']>= 1880) & (h['YearBuilt']< 1890)), 'Year']= '1880s'
h.loc[((h['YearBuilt']>= 1890) & (h['YearBuilt']< 1900)), 'Year']= '1890s'
h.loc[((h['YearBuilt']>= 1900) & (h['YearBuilt']< 1910)), 'Year']= '1900s'
h.loc[((h['YearBuilt']>= 1910) & (h['YearBuilt']< 1920)), 'Year']= '1910s'
h.loc[((h['YearBuilt']>= 1920) & (h['YearBuilt']< 1930)), 'Year']= '1920s'
h.loc[((h['YearBuilt']>= 1930) & (h['YearBuilt']< 1940)), 'Year']= '1930s'
h.loc[((h['YearBuilt']>= 1940) & (h['YearBuilt']< 1950)), 'Year']= '1940s'
h.loc[((h['YearBuilt']>= 1950) & (h['YearBuilt']< 1960)), 'Year']= '1950s'
h.loc[((h['YearBuilt']>= 1960) & (h['YearBuilt']< 1970)), 'Year']= '1960s'
h.loc[((h['YearBuilt']>= 1970) & (h['YearBuilt']< 1980)), 'Year']= '1970s'
h.loc[((h['YearBuilt']>= 1980) & (h['YearBuilt']< 1990)), 'Year']= '1980s'
h.loc[((h['YearBuilt']>= 1990) & (h['YearBuilt']< 2000)), 'Year']= '1990s'
h.loc[((h['YearBuilt']>= 2000) & (h['YearBuilt']< 2010)), 'Year']= '2000s'
h.loc[((h['YearBuilt']>= 2010) & (h['YearBuilt']< 2020)), 'Year']= '2010s'
```
